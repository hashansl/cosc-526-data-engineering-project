TDA Mapper

```python
# Data viz
from gtda.plotting import plot_point_cloud

# TDA magic
from gtda.mapper import (
    CubicalCover,
    make_mapper_pipeline,
    Projection,
    plot_static_mapper_graph,
    plot_interactive_mapper_graph,
)
# from gtda.mapper.utils.visualization import set_node_sizeref
```

```python
# ML tools
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
from gtda.mapper.filter import FilterFunctionName
from gtda.mapper.cover import CoverName
# scikit-learn method
from sklearn.cluster import ClusteringAlgorithm
# giotto-tda method
from gtda.mapper.cluster import FirstSimpleGap
```

```python
# Define filter function – can be any scikit-learn transformer
filter_func = Projection(columns=[0, 1])
```

```python
# Define cover
cover = CubicalCover(n_intervals=10, overlap_frac=0.3)
```

```python
# Choose clustering algorithm – default is DBSCAN
clusterer = DBSCAN()
```

```python
# Configure parallelism of clustering step
n_jobs = 1
```

```python
# Initialise pipeline
pipe = make_mapper_pipeline(
    filter_func=filter_func,
    cover=cover,
    clusterer=clusterer,
    verbose=False,
    n_jobs=n_jobs,
)
```

```python
parts['avg_embedding'][0]
```

```python
# Convert to 2D NumPy array
X = np.vstack(parts['avg_embedding'].values)
```

```python
X.shape
```

```python
# Step 1: Lens (projection to 2D using PCA)
lens = PCA(n_components=2)
```

```python
# Step 2: Cover definition
cover = CubicalCover(n_intervals=10, overlap_frac=0.2)
```

```python
# Step 3: Clustering algorithm
clusterer = DBSCAN(eps=0.5, min_samples=5)
```

```python
# Step 4: Build the Mapper pipeline
mapper_pipeline = make_mapper_pipeline(
    filter_func=lens,
    cover=cover,
    clusterer=clusterer
)
```

```python

# Step 5: Fit the pipeline
graph = mapper_pipeline.fit_transform(X)
```

```python
fig = plot_static_mapper_graph(mapper_pipeline, X, layout_dim=2)
fig.show()
```

```python
fig = plot_static_mapper_graph(
    mapper_pipeline,
    X,
    color_data=parts['cluster'].values,  # or 'title', etc.
    layout_dim=2
)
fig.show()
```
